```python
highlight_classes = {'Calanoid_1', 'Cyclopoid_1', 'Bosmina_1', 'Herpacticoida', 'Chironomid', 'Chydoridae', 'Daphnia'}

# Folder containing CSV files
csv_folder = './Plankton/SIMC.Overlap.csv/'

# Base directory containing image files
image_base_dir = './Plankton/SIMC_OverlapTiffsWithPP/SIMC_OverlapTiffsWithPP/'

# Output directory for extracted snippets
output_dir = 'extracted_snippets'
os.makedirs(output_dir, exist_ok=True)

# Iterate through all CSV files in the folder
for csv_file in glob.glob(os.path.join(csv_folder, '*.csv')):
    df = pd.read_csv(csv_file)
    csv_basename = os.path.splitext(os.path.basename(csv_file))[0]

    # Iterate through each record
    for idx, row in df.iterrows():
        if row['Class'] in highlight_classes:
            image_path = os.path.join(image_base_dir, row['Image.File'])
            x, y = row['Image.X'], row['Image.Y']
            width, height = row['Image.Width'], row['Image.Height']

            # Load image
            with Image.open(image_path) as img:
                # Define the box to crop
                box = (x, y, x + width, y + height)
                snippet = img.crop(box)

                # Save snippet with a meaningful name
                snippet_filename = f"{csv_basename}_Particle_{row['Particle.ID']}_{row['Class']}.png"
                snippet_filepath = os.path.join(output_dir, snippet_filename)
                snippet.save(snippet_filepath)

print(f"Extraction completed. Snippets saved to '{output_dir}' directory.")
```
